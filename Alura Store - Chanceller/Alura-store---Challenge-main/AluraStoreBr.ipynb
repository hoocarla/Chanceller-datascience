#1. Análise do faturamento


```python
#data frames que contem as informações de cada loja
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
df4 = pd.read_csv(url4)

# Calcular faturamento total de cada loja
faturamento_1 = df1['Preço'].sum()
faturamento_2 = df2['Preço'].sum()
faturamento_3 = df3['Preço'].sum()
faturamento_4 = df4['Preço'].sum()

# Exibir os resultados
print(f"Faturamento total da Loja 1: R${faturamento_1:.2f}")
print(f"Faturamento total da Loja 2: R${faturamento_2:.2f}")
print(f"Faturamento total da Loja 3: R${faturamento_3:.2f}")
print(f"Faturamento total da Loja 4: R${faturamento_4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Criar um dicionário com os dados
faturamento_lojas = {
    'Loja 1': faturamento_1,
    'Loja 2': faturamento_2,
    'Loja 3': faturamento_3,
    'Loja 4': faturamento_4,
}

# Criar o gráfico de barras
plt.figure(figsize=(15, 7))
bars = plt.bar(faturamento_lojas.keys(), faturamento_lojas.values(), color='slateblue', edgecolor='black', linewidth=1)
plt.xticks(rotation=45)

# Adicionar título e rótulos
plt.gca().yaxis.set_major_formatter('R${x:.2f}')
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)', fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Objeto de eixo
ax = plt.gca()

# Adicionar rótulos de dados em cima de cada barra
ax.bar_label(bars, labels=[f'R${valor:,.2f}' for valor in faturamento_lojas.values()], padding=3)

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
#criaçao de um dicionario para armazenar os dados das 4 lojas
lojas = {
    'Loja 1': df1,
    'Loja 2': df2,
    'Loja 3': df3,
    'Loja 4': df4
}

for nome_loja, df in lojas.items():
    print(f'\n📊 Categorias mais vendidas - {nome_loja}')
    categorias = df['Categoria do Produto'].value_counts()
    total_vendas = categorias.values.sum()
    print(f"Total de itens vendidos em todas as categorias: {loja} {total_vendas} ")
    print(categorias)

# Criando subplots
fig, axs = plt.subplots(2, 2, figsize=(18, 9))
fig.suptitle('Categorias Mais Vendidas por Loja', fontsize=16)

# Loop para plotar cada gráfico
for i, (nome_loja, df) in enumerate(lojas.items()):
    categorias = df['Categoria do Produto'].value_counts()

    ax = axs[i // 2, i % 2]
    bars = ax.bar(categorias.index, categorias.values, color='skyblue')

    ax.set_title(nome_loja)
    ax.set_ylabel('Quantidade Vendida')
    ax.set_xlabel('Categoria')
    ax.tick_params(axis='x', rotation=45)

    # Adiciona rótulos de dados em cima de cada barra
    ax.bar_label(bars, labels=categorias.values, padding=3)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
#importar a biblioteca para realizar a funcao de media
import statistics

#criaçao de um dicionario para armazenar os dados das 4 lojas
lojas = {
    'Loja 1': df1,
    'Loja 2': df2,
    'Loja 3': df3,
    'Loja 4': df4
}

# Calcular a media de avaliações de cada loja
media_1 = df1['Avaliação da compra'].mean()
media_2 = df2['Avaliação da compra'].mean()
media_3 = df3['Avaliação da compra'].mean()
media_4 = df4['Avaliação da compra'].mean()

# Exibir os resultados
print(f"A media das avaliações da Loja 1: {media_1:.2f}")
print(f"A media das avaliações da Loja 1: {media_2:.2f}")
print(f"A media das avaliações da Loja 1: {media_3:.2f}")
print(f"A media das avaliações da Loja 1: {media_4:.2f}")

# Nomes das lojas e médias correspondentes
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = [media_1, media_2, media_3, media_4]

# Criando o gráfico
plt.figure(figsize=(8, 5))
plt.plot(lojas, medias, marker='o', linestyle='-', color='green', linewidth=2)

#rotulo dos dados
for loja, media in zip(lojas, medias):
    plt.text(loja, media, f'{media:.2f}', ha='center', va='bottom')

# Título e rótulos
plt.title('Média das Avaliações por Loja', fontsize=14)
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.ylim(0, 5)  # Supondo que a avaliação vai de 0 a 5

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
import matplotlib.pyplot as plt


# Criar um dicionario para todas as lojas
lojas = {
    'Loja 1': df1,
    'Loja 2': df2,
    'Loja 3': df3,
    'Loja 4': df4
}

#repetição para fazer a leitura de valores de cada loja mais vendido
for nome_loja, df in lojas.items():
    # Encontrar o produto mais vendido para cada loja
    maior_venda = df['Produto'].value_counts().idxmax()
    quantidade_vendida = df['Produto'].value_counts().max()
    print(f"O Produto mais vendido da {nome_loja} é: {maior_venda} com {quantidade_vendida} vendas")
```

```python
#Gerar o Grafico

import matplotlib.pyplot as plt

# Criar listas para armazenar os dados do loop
nomes_lojas = []
quantidades = []
produtos = []

# Reexecutar o loop apenas para armazenar os dados
for nome_loja, df in lojas.items():
    maior_venda = df['Produto'].value_counts().idxmax()
    quantidade_vendida = df['Produto'].value_counts().max()

#Para adicionar um elemento ao final da lista e exibir as 4 lojas no grafico
    nomes_lojas.append(nome_loja)
    quantidades.append(quantidade_vendida)
    produtos.append(maior_venda)

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
barras = plt.bar(nomes_lojas, quantidades, color='steelblue')

# Adicionar rótulos nas barras (produto + quantidade)
for i, barra in enumerate(barras):
    altura = barra.get_height()
    rotulo = f"{produtos[i]}\n({quantidades[i]})"
    plt.text(barra.get_x() + barra.get_width()/2, altura + 1, rotulo, ha='center', va='bottom', fontsize=9, fontweight='bold')

# Títulos e ajustes
plt.title('Produto Mais Vendido por Loja')
plt.xlabel('Lojas')
plt.ylabel('Quantidade Vendida')
plt.ylim(0, max(quantidades) + 10)
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
#importar a biblioteca para realizar a funcao de media
import statistics, matplotlib.pyplot as plt

#criaçao de um dicionario para armazenar os dados das 4 lojas
lojas = {
    'Loja 1': df1,
    'Loja 2': df2,
    'Loja 3': df3,
    'Loja 4': df4
}

#laço de repetição for para fazer a media das 4 lojas
for nome_loja, df in lojas.items():
    media_frete = df['Frete'].mean()
    print(f"O frete médio da {nome_loja} é: R$ {media_frete:.2f}")




# Extrair e ordenar lojas e médias
lojas_ordenadas = sorted(lojas.keys(), key=lambda x: int(x.split()[1]))
medias = [df['Frete'].mean() for df in lojas.values()]

# Tamanho do Grafíco
plt.figure(figsize=(15, 8))
plt.plot(lojas_ordenadas, medias, marker='o', color='olive')

# Adicionar rótulos
for loja, media in zip(lojas_ordenadas, medias):
    plt.text(loja, media, f'R${media:.2f}', ha='center', va='bottom')

# Configurações básicas
plt.gca().yaxis.set_major_formatter('R${x:.2f}')
plt.ylabel('Valor do Frete')
plt.title('Média de Frete por Loja')
plt.grid(True)
plt.show()
```

6. GRÁFICO DE LONGETITUDE E LATITUDE

```python
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt

# Vendas por cidade de cada loja
vendas_cidade_1 = df1['Local da compra'].value_counts()
vendas_cidade_2 = df2['Local da compra'].value_counts()
vendas_cidade_3 = df3['Local da compra'].value_counts()
vendas_cidade_4 = df4['Local da compra'].value_counts()

# Juntar tudo em um DataFrame
vendas_cidades = pd.concat([vendas_cidade_1, vendas_cidade_2, vendas_cidade_3, vendas_cidade_4], axis=1)

# Renomear as colunas
vendas_cidades.columns = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# # Visualizar as informações e arredondar as casas decimais
# print(vendas_cidades.to_string(formatters={'Loja 1': '{:.0f}'.format, 'Loja 2': '{:.0f}'.format, 'Loja 3': '{:.0f}'.format, 'Loja 4': '{:.0f}'.format}))
# vendas_cidades = vendas_cidades.fillna(0)

#gerar um grafico de barras para ter uma noção das cidades mais vendidas em cada loja

# Criar o gráfico de barras
import matplotlib.pyplot as plt

# Lista de lojas
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Criar a figura e os eixos (2 linhas, 2 colunas)
fig, axs = plt.subplots(2, 2, figsize=(20, 10))  # aumenta o tamanho da figura

# Deixar espaço entre os gráficos
fig.tight_layout(pad=5.0)

# Transformar o array 2x2 em lista para usar no for
axs = axs.flatten()

# Criar um gráfico para cada loja
for i, loja in enumerate(lojas):
    axs[i].bar(vendas_cidades.index, vendas_cidades[loja], color='purple', edgecolor='black')
    axs[i].set_title(f'{loja}', fontsize=14)
    axs[i].set_xlabel('Estados', fontsize=12)
    axs[i].set_ylabel('Quantidade Vendida', fontsize=12)
    axs[i].tick_params(axis='x', rotation=45)

    # Adicionar rótulos nas barras
    for bar in axs[i].patches:
        altura = bar.get_height()
        # Check if the height is NaN and replace it with 0
        if pd.isnull(altura):
            altura = 0
        axs[i].text(bar.get_x() + bar.get_width()/2, altura + 5,
                    f'{int(altura)}', ha='center', va='bottom', fontsize=9) # Convert to int after checking for NaN



# Adicionar um título geral no meio
fig.suptitle('Quantidade de vendas por Estado', fontsize=20)

plt.show()

```

```python
# data frame para saber os dados de cada cidade
dados_cidade_1 = df1[['Local da compra', 'lat', 'lon']]
dados_cidade_2 = df2[['Local da compra', 'lat', 'lon']]
dados_cidade_3 = df3[['Local da compra', 'lat', 'lon']]
dados_cidade_4 = df4[['Local da compra', 'lat', 'lon']]

# Juntar tudo em um DataFrame
dados_cidades = pd.concat([dados_cidade_1, dados_cidade_2, dados_cidade_3, dados_cidade_4])


# Agrupar por cidade e pegar a primeira lat/lon de cada cidade
coordenadas_cidades = dados_cidades.groupby('Local da compra').agg({'lat':'mean', 'lon':'mean'}).reset_index()
coordenadas_cidades
```
